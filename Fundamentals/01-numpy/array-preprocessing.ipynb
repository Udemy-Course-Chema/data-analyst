```python
import numpy as np
import matplotlib.pyplot as plt

# To see visualization in IPython.
%matplotlib inline
```

```python
points = np.arange(-5, 5,0.01)
points
```

Coordinate Vectors

```python
dx, dy = np.meshgrid(points, points)
```

```python
dx
```

```python
dy
```

```python
z = ( np.sin(dx) + np.sin(dy)  )
z
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()

plt.title('Plot for sin(x)+sin(z)')
```

Numpy

```python
A = np.array([1,2,3,4])
B = np.array([100,200,300,400])
```

```python
condition = np.array([True, True, False, False] )
```

```python
answer = [ (A_val if cond else B_var) for A_val, B_var, cond in zip(A,B,condition) ]
answer
```

```python
answer2 = np.where(condition, A,B )
answer2
```
